```python
from bs4 import BeautifulSoup 
import requests as rq
import pandas as pd
import openpyxl
```

```python
name= []
price=[]
mileage=[]
dealer_name=[]
rating_count=[]
rating=[]

for i in range(1,11):
    url='https://www.cars.com/shopping/results/?list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&page=' +str(i)+'&stock_type=cpo&zip='
    response= rq.get(url)
    soup= BeautifulSoup(response.content, 'html.parser')
    res= soup.find_all('div',{'class':'vehicle-card-main js-gallery-click-card'})
   
    for result in res:
        try: 
            name.append(result.find('h2').get_text())   
        except:
            name.append('n/a')

        try:
            price.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            price.append('n/a')
        
        try:
            mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')

        try:
            dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')
            
        try:
            rating_count.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            rating_count.append('n/a')

        try:
            rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')  
```

```python
dealer= pd.DataFrame({'Name':name, 'Price':price, 'Mileage': mileage, 'Dealer_Name':dealer_name, 'Rating': rating, 'Rating_Count':rating_count})
dealer.head(20)
```

```python
dealer['Rating_Count']= dealer['Rating_Count'].apply(lambda x:x.strip('reviews)').strip('(') )
dealer.head()
```

```python
dealer.to_excel('multipage_page_car.xlsx', index=False)
```

```python
from bs4 import BeautifulSoup 
import requests as rq
import sqlite3

# Create SQLite database connection
conn = sqlite3.connect('car_data.db')
cursor = conn.cursor()

# Create car_data table if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS car_data (
        Name TEXT,
        Price TEXT,
        Mileage TEXT,
        Dealer_Name TEXT,
        Rating TEXT,
        Rating_Count TEXT
    )
''')
```

```python
# Prepare SQL statement for inserting data
insert_query = 'INSERT INTO car_data (Name, Price, Mileage, Dealer_Name, Rating, Rating_Count) VALUES (?, ?, ?, ?, ?, ?)'

name= []
price=[]
mileage=[]
dealer_name=[]
rating_count=[]
rating=[]

for i in range(1,11):
    url='https://www.cars.com/shopping/results/?list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&page=' +str(i)+'&stock_type=cpo&zip='
    response= rq.get(url)
    soup= BeautifulSoup(response.content, 'html.parser')
    res= soup.find_all('div',{'class':'vehicle-card-main js-gallery-click-card'})
   
    for result in res:
        try: 
            name.append(result.find('h2').get_text())   
        except:
            name.append('n/a')

        try:
            price.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            price.append('n/a')
        
        try:
            mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')

        try:
            dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')
            
        try:
            rating_count.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            rating_count.append('n/a')

        try:
            rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')  
     # Insert data into SQLite database
    data = list(zip(name, price, mileage, dealer_name, rating, rating_count))
    cursor.executemany(insert_query, data)
    conn.commit()

# Close the database connection
cursor.close()
conn.close()
```
